### Задание №2

```python
def create_plot(func, start_x=-100.0, stop_x=100.0, step_x=0.1):
  x = np.arange(start_x, stop_x+step_x, step_x)
  y = func(x)
  plt.plot(x, y)
  plt.show()
```

Построить график функци:

1. $y=x^2+4⋅x+3$;

```python
def y1(x):
  return x*x+4*x+3

create_plot(y1)
```

2. $y=-2⋅\sin{(3⋅x)};$

```python
def y2(x):
  return -2*np.sin(3*x)

create_plot(y2, -10,10)
```

3. $y=|\{x\}-\frac{1}{2}|$.

```python
def y3(x):
  result = np.zeros(x.shape)
  for inx, el in enumerate(x):
    result[inx] = np.abs(np.transpose(np.modf(el))[0]-0.5)
  return result

create_plot(y3, -10,10)
```

### Задание №3

Найти пределы:

1. $\lim_{x \to 5}{\frac{x^2-6x+5}{x^2-25}}$;

$$\lim_{x \to 5}{\frac{x^2-6x+5}{x^2-25}}=\lim_{x \to 5}{\frac{(x-1)⋅(x-5)}{(x-5)⋅(x+5)}}=\lim_{x \to 5}{\frac{x-1}{x+5}}=\frac{4}{10}=0.4$$

2. $\lim_{x \to -1}{\frac{x^3+x+2}{x^3+1}}$;

$$\lim_{x \to -1}{\frac{x^3+x+2}{x^3+1}}=\lim_{x \to -1}{\frac{(x+1)⋅(x^2-x+2)}{(x+1)⋅(x^2-x+1)}}=\lim_{x \to -1}{\frac{x^2-x+2}{x^2-x+1}}=\frac{1+1+2}{1+1+1}=\frac{4}{3}$$

3. $\lim_{x \to 3}{\frac{\sqrt{2⋅x+3}-3}{\sqrt{x-2}-1}}$;

$$\lim_{x \to 3}{\frac{\sqrt{2⋅x+3}-3}{\sqrt{x-2}-1}}=\lim_{x \to 3}{\frac{(2⋅x+3-9)⋅(\sqrt{x-2}+1)}{(x-2-1)⋅(\sqrt{2⋅x+3}+3)}}=\lim_{x \to 3}{\frac{2⋅(x-3)⋅(\sqrt{x-2}+1)}{(x-3)⋅(\sqrt{2⋅x+3}+3)}}=\frac{2⋅(\sqrt{3-2}+1)}{\sqrt{2⋅3+3}+3}=\frac{4}{6}=\frac{2}{3}$$

4. $\lim_{x \to 0}{\frac{1-\cos{x}}{x^2}}$;

$$\lim_{x \to 0}{\frac{1-\cos{x}}{x^2}}=\lim_{x \to 0}{\frac{(1-\cos{x})⋅(1+\cos{x})}{x^2⋅(1+\cos{x})}}=\lim_{x \to 0}{\frac{1-(\cos{x})^2}{x^2⋅(1+\cos{x})}}=\lim_{x \to 0}{\frac{(\sin{x})^2}{x^2⋅(1+\cos{x})}}=\frac{1}{2}$$

5. $\lim_{x \to 0}{x⋅\cot{x}}$;

$$\lim_{x \to 0}{x⋅\cot{x}}=\lim_{x \to 0}{\frac{x⋅\cos{x}}{\sin{x}}}=1$$

6. $\lim_{x \to 0}{\sqrt[2⋅x]{1+3⋅x}}$;

$$\lim_{x \to 0}{\sqrt[2⋅x]{1+3⋅x}}=\lim_{x \to 0}{(1+3⋅x)^{\frac{1}{2⋅x}}}=\lim_{x \to 0}(1+3⋅x)^{\frac{3}{(3⋅x)⋅2}}=\begin{vmatrix}y=3⋅x \\ x \to 0 => y\to 0 \end{vmatrix}=\lim_{x \to 0}{(1+y)^{\frac{3}{2⋅y}}}=e^{\frac{3}{2}}$$

7. $\lim_{x \to 0}{(\frac{3+5⋅x}{3+2⋅x})^{\frac{1}{x}}}$.

$$\lim_{x \to 0}{(\frac{3+5⋅x}{3+2⋅x})^{\frac{1}{x}}}=\lim_{x \to 0}{(\frac{1+\frac{5}{3}⋅x}{1+\frac{2}{3}⋅x})^{\frac{1}{x}}}=\lim_{x \to 0}{\frac{(1+\frac{5}{3}⋅x)^{\frac{5⋅3}{3⋅x⋅5}}}{(1+\frac{2}{3}⋅x)^{\frac{2⋅3}{2⋅x⋅3}}}}=\frac{e^{\frac{3}{5}}}{e^{\frac{3}{2}}}=e^{\frac{3}{5}-\frac{3}{2}}=e^{-\frac{9}{10}}$$
